Problems in kmeans algo  - cant work over nonlinear data / complex data

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons

x,y = make_moons(n_samples=200,noise=0.10)
```

```python
plt.scatter(x[:,0], x[:, 1], c=y)
plt.show()
```

```python
import os
os.environ["OMP_NUM_THREADS"] = "1"
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 2)
km.fit(x)
```

```python
centers = km.cluster_centers_
label = km.labels_
```

```python
plt.scatter(x[:,0], x[:,1], c=label)
plt.scatter(centers[:,0], centers[:,1], color='red')
plt.show()
```

Density based spatial clustering of applications with noise. 
    Based upon on the idea, that a cluster is a high density area surrounded by low density area.
    Starts by exploring a small area , if density is "good enough" it considered as part of cluster and explores neighbors and  so on, so it groups regions based upon density
